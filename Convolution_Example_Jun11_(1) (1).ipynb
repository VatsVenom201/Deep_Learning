## Convolution in Real Images

- #### Edge Detector

## Step 1: Import the Necessary Packages
- Download the image, convert it to a numpy array, and then read it into OpenCV format
- Return the image

```python
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

def url_to_image(url):
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)

	return image
```

## Step 2: Save the Image Array in a Variable
- Load the image from the URL
- Retrieve the shape of the image
- Display the image


```python
image = url_to_image("https://upload.wikimedia.org/wikipedia/commons/0/09/TheCheethcat.jpg")
image.shape
```

```python
plt.imshow(image)
```

## Step 3: Vertical Edge Detector
- To detect vertical edges we can convolve our image to a vertical edge detection kernel as: $\begin{bmatrix}-1 & 0 & 1\\-1 & 0 & 1\\-1 & 0 & 1\end{bmatrix}$  which is going to give us an output image with only vertical edges.
-Let's examine the image to identify any vertical edges present in the image.

```python
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
plt.imshow(dst)
```

##Step 4: Horizontal Edge Detector

- To detect horizontal edges we can convolve our image to a horizontal edge detector kernel as $\begin{bmatrix}-1 & -1 & -1\\ 0 & 0 & 0\\1 & 1 & 1\end{bmatrix}$ to get an image with only horizontal edges.
- The image below displays the output of a horizontal edge detector.


```python
kernel = np.float32([[-1,-1,-1],[0,0,0],[1,1,1]])

dst = cv2.filter2D(image,-1,kernel)
plt.imshow(dst)
```

## Step 5: Sharpened Kernel
- To enhance the sharpness of the original image, we can apply a sharpening kernel as $\begin{bmatrix}-1 & -1 & -1\\ -1 & 9 & -1\\-1 & -1 & -1\end{bmatrix}$ which is going to give us sharper version of the image.
- Let's examine the output image obtained by applying the sharpening kernel for image enhancement.

```python
kernel = np.float32([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])

dst = cv2.filter2D(image,-1,kernel)
plt.imshow(dst)
```
